```python
from datasets import load_dataset
import pandas as pd

# Load all four NLI tasks
mnli   = load_dataset("IRIIS-RESEARCH/MNLI-Nepali")
qnli   = load_dataset("IRIIS-RESEARCH/QNLI-Nepali")
rte    = load_dataset("IRIIS-RESEARCH/RTE-Nepali")
coref  = load_dataset("IRIIS-RESEARCH/Co-Reference-Nepali")
xnli   = load_dataset("IRIIS-RESEARCH/XNLI-Nepali")

# Check structure of MNLI
print(mnli)
print(mnli['train'][0])          # see column names and a sample
print(mnli['train'].features)    # see label encoding

# Label distribution
import collections
labels = mnli['train']['label']  # or whatever the column is called
print(collections.Counter(labels))
```

```python
print(collections.Counter(mnli['test']['label']))
```
